```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.inspection import permutation_importance
from imblearn.combine import SMOTETomek

# Load the dataset
file_path = 'train.csv'  # Update this path as needed
df = pd.read_csv(file_path)

# Drop irrelevant columns
columns_to_drop = ['physical_part_type', 'message_timestamp', 'weekday', 'shift', 'physical_part_id']
df = df.drop(columns=columns_to_drop, errors='ignore')

# Encode 'status' column using LabelEncoder
label_encoder = LabelEncoder()
df['status_encoded'] = label_encoder.fit_transform(df['status'])

# Convert columns to numeric where applicable, except for 'status'
columns_to_exclude = ['status', 'status_encoded']
columns_to_convert = [col for col in df.columns if col not in columns_to_exclude]
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Handle missing values: Fill with median or interpolate
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.fillna(x.median()) if x.isna().sum() > 0 else x)

# Outlier detection using IQR
Q1 = df[columns_to_convert].quantile(0.25)
Q3 = df[columns_to_convert].quantile(0.75)
IQR = Q3 - Q1

# Filtering outliers outside 1.5*IQR
df_filtered = df[~((df[columns_to_convert] < (Q1 - 1.5 * IQR)) | (df[columns_to_convert] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(f"After outlier removal: {df_filtered.shape[0]} rows remain.")

# Aggregating sensor data (mean, max, min) across all columns related to sensors
sensor_columns = [col for col in df_filtered.columns if 'sensor' in col]

aggregated_df = pd.concat([
    df_filtered,
    df_filtered[sensor_columns].mean(axis=1).rename('sensor_mean'),
    df_filtered[sensor_columns].max(axis=1).rename('sensor_max'),
    df_filtered[sensor_columns].min(axis=1).rename('sensor_min')
], axis=1)

# Feature reduction using Variance Threshold
selector = VarianceThreshold(threshold=0.1)  # You can adjust this threshold
reduced_df = selector.fit_transform(aggregated_df[columns_to_convert])
reduced_df = pd.DataFrame(reduced_df, columns=[columns_to_convert[i] for i in range(len(columns_to_convert)) if selector.get_support()[i]])

# Standardizing the features before PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(reduced_df)

# Applying PCA to reduce dimensionality
pca = PCA(n_components=10)  # You can adjust the number of components
pca_features = pca.fit_transform(scaled_data)

# Adding PCA features to the dataframe
pca_df = pd.DataFrame(pca_features, columns=[f'pca_feature_{i+1}' for i in range(pca_features.shape[1])])
final_df = pd.concat([reduced_df.reset_index(drop=True), pca_df, aggregated_df[['status_encoded']].reset_index(drop=True)], axis=1)

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_}")

# Display the shape of the cleaned dataset
print(f"Cleaned dataset contains {final_df.shape[0]} rows and {final_df.shape[1]} columns after data cleaning and feature engineering.")

# Save the cleaned dataset to a CSV file
output_file_path = 'cleaned_dataset4.csv'  # Update this path as needed
final_df.to_csv(output_file_path, index=False)
print(f"Cleaned dataset saved to {output_file_path}")

# Define features and target
X = final_df.drop(columns=['status_encoded'])
y = final_df['status_encoded']

# Handle class imbalance using SMOTETomek
smote_tomek = SMOTETomek(random_state=42)
X_resampled, y_resampled = smote_tomek.fit_resample(X, y)

# Split the resampled data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")

# Train a RandomForest model to use permutation feature importance analysis
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)

# Use permutation importance to analyze feature importance
perm_importance = permutation_importance(rf_clf, X_train, y_train, n_repeats=10, random_state=42)
perm_importances = pd.Series(perm_importance.importances_mean, index=X_train.columns)

# Plot feature importance
plt.figure(figsize=(10, 6))
perm_importances.nlargest(20).plot(kind='barh')
plt.title('Top 20 Features by Permutation Importance')
plt.show()

# Select top features based on permutation importance
top_features = perm_importances.nlargest(20).index
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]

# Define models for stacking
xgb_clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
lgb_clf_top = LGBMClassifier(random_state=42)

# Create a stacking classifier
estimators = [
    ('rf', rf_clf),
    ('xgb', xgb_clf),
    ('lgb', lgb_clf_top)
]
stacking_clf = StackingClassifier(estimators=estimators, final_estimator=LGBMClassifier(random_state=42))

# Train and evaluate the stacking model
print("\nTraining Stacking Classifier...")
stacking_clf.fit(X_train_top, y_train)
y_pred = stacking_clf.predict(X_test_top)

# Evaluationa
print(f"\nStacking Classifier - Classification Report:")
print(classification_report(y_test, y_pred))
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred)}")
```
